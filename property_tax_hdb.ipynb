```python
print(df.type_of_hdb.unique())
toh=input('Enter Type of Hdb:')
Type=df[df.type_of_hdb == toh].reset_index()
Type
```

```python
plt.figure(figsize=(20,6))
sb.lineplot(x=Type.financial_year,y=Type.no_of_cases, color='#243848')
plt.title('No of Cases over the years')
plt.xlabel('Financial Year')
plt.ylabel('No of Cases')
plt.tick_params(axis='x', rotation=90)
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20, 15))

sb.barplot(x=Type.financial_year, y=Type.median_annual_value, color='#243848',edgecolor='grey', ax=axs[0])
axs[0].set_title('Median annual value over the years')
axs[0].set_xlabel('Financial Year')
axs[0].set_ylabel('Median Annual Value')
axs[0].tick_params(axis='x', rotation=90)

sb.barplot(x=Type.financial_year, y=Type.property_tax_collection, color='#243848',edgecolor='grey', ax=axs[1])
axs[1].set_title('Property Tax Collection over the years')
axs[1].set_xlabel('Financial Year')
axs[1].set_ylabel('Property Tax Collection')
axs[1].tick_params(axis='x', rotation=90)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sb.boxplot(x='type_of_hdb', y='median_annual_value',color='#243848',data=Type)
plt.title('Distribution of Median Annual Value by Housing Type')
plt.xlabel('Housing Type')
plt.ylabel('Median Annual Value')

plt.subplot(1, 2, 2)
sb.boxplot(x='type_of_hdb', y='property_tax_collection',color='#243848', data=Type)
plt.title('Distribution of Property Tax Collection by Housing Type')
plt.xlabel('Housing Type')
plt.ylabel('Property Tax Collection')

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[['no_of_cases','median_annual_value','property_tax_collection']].corr()

plt.figure(figsize=(10, 8))
sb.heatmap(corr_matrix, annot=True, cmap='Blues_r', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Heatmap')
plt.show()
```
